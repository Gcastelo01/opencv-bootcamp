# Primeira aula de OpenCV

NEsta aula, iremos aprender mais sobre como trabalhar com imagens usando a biblioteca OpenCV em python. 

````python
# Importando Bibliotecas

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
````

````python
# Mostrando imagens com Ipython:

Image(filename="./images/friends.jpeg")
````

````python
# Agora, vmaos usar o opencv para trabalhar:

img = cv.imread('./images/friends.jpeg', 0)

plt.imshow(img, cmap='gray')
````

````python
# Agora, com cores!

# Separando componentes (cv le as imagens em bgr, plt em rgb)
img = cv.imread('./images/friends.jpeg', cv.IMREAD_COLOR)
b,g,r = cv.split(img)

plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(r, cmap='gray');plt.title("redChannel")
plt.subplot(142); plt.imshow(g, cmap='gray');plt.title("greenChannel")
plt.subplot(143); plt.imshow(b, cmap='gray');plt.title("blueChannel")

plt.show()
````

````python
# Agora, vamos converter para o canal RGB usando cv.cvtColor

im = cv.imread("./images/friends.jpeg", cv.IMREAD_COLOR)
im_cv = cv.cvtColor(im, cv.COLOR_BGR2RGB)

plt.figure(figsize=[10, 5])
plt.subplot(121); plt.imshow(im); plt.title("Imagem sem conversão")
plt.subplot(122); plt.imshow(im_cv); plt.title("Imagem com conversão")

plt.show()
````

## O que foi visto até agora?

Vimos até então como abrir uma imagem utilizando openCV, como podemos plotar essa imagem com apoio do plt e como converter o formato da imagem para outros. 

## O que veremos?
Como podemos modificar canais individuais:

````python
# Separamos os canais:
im_cv = cv.imread('./images/friends.jpeg')
im_cv =cv.cvtColor(im_cv, cv.COLOR_BGR2RGB)

r, g, b = cv.split(im_cv)

r_new = r+120

im_merged = cv.merge((r_new,g,b))

plt.imshow(im_merged)
````
